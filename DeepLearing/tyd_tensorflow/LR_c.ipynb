## 做一个线性回归的列子

```python
#　引入相应的模块
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
# 使用numpy产生100个随机点,是线的围绕在 y=0.1x + 0.3 的附近
num_point = 100
vectors_set = []
for i in range(num_point):
    x = np.random.normal(0.0, 0.5)
    y = x *0.1 + 0.3 + np.random.normal(0.0, 0.03)  # 在这个增加一个偏差
    vectors_set.append([x, y])


x_data = [v[0] for v in vectors_set]
y_data = [v[1] for v in vectors_set]
# 使用点来显示 
# tf.summary.scatter(vectors_set, name="point")

# 使用plt显示图片
plt.scatter(x_data, y_data, color='red')
plt.show()
```

```python
# 生成一个一维的W矩阵, 取值是-1到1的一个随机值 (weight)
with tf.name_scope("weight"):
    W = tf.Variable(tf.random_uniform([1], -1.0, 1.0 ), name="W")
# 生成一个一维的b矩阵, 初始化值为0  (bias)
with tf.name_scope("bias"):
    b = tf.Variable(tf.zeros([1], name="B")) 

# 计算出预测值 y_pre
with tf.name_scope("y"):
    y_pre = W * x_data + b
# print(y_pre)

#  计算实际值 y_data 和 预测值 y_pre 之间的损失,,使用均方差作为损失. 先相减,然后平方,然后平均值
with tf.name_scope("loss"):
    loss = tf.reduce_mean(tf.square(y_data - y_pre), name="loss")

# 使用梯度下降来优化参数

optimizer = tf.train.GradientDescentOptimizer(0.5)
# 训练的过程就是要使误差值最小化 
train = optimizer.minimize(loss, name="trian")

# 初始化数据
init = tf.global_variables_initializer()

# 建立会话(session)
with tf.Session() as sess:
    sess.run(init)
    print(sess.run(W))
    print("W :",sess.run(W),"* b :",sess.run(b), "* loss :",sess.run(loss))
    print("^"*20)
    # 训练20次
    for i in range(21):
        sess.run(train)
        print("W :",sess.run(W),"* b :",sess.run(b), "* loss :",sess.run(loss))
```

```python
# 可视化, 使用tensorboard
writer = tf.summary.FileWriter("logs/", sess.graph)
```
