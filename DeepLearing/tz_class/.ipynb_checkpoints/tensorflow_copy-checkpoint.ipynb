```python
import tensorflow as tf
```

```python
# 定义二个常量 op
# with tf.name_scope as 
m1 = tf.constant([[3,4]], tf.float32)
m2 = tf.constant([[5],[6]], tf.float32)
# 创建一个矩阵乘法op, 并讲m1,m2传入
product = tf.matmul(m1, m2)

#定义会话
sess = tf.Session()
# 使用会话
result = sess.run(product)

print(sess.run(product))
sess.close()
# 或者使用with来定义会话.
# with tf.Session() as sess:
#     sess.run(product)
```

```python
x = tf.Variable([1,2])
a = tf.constant([3,3])

sub = tf.subtract(x,a)
add = tf.add(x, sub)

init = tf.global_variables_initializer()

with tf.Session() as sess:
    sess.run(init)
    sess.run(sub)
    sess.run(add)
    print(sess.run(sub))
    print(sess.run(add))
```

```python
state = tf.Variable(0, name='counter')
new_value = tf.add(state, 1)
# 赋值op
update = tf.assign(state, new_value)

init = tf.global_variables_initializer()

with tf.Session() as sess:
    sess.run(init)
    print(sess.run(state))
    for _ in range(10):
        sess.run(update)
        print(sess.run(state))
```

```python
# 传值和提取值


import tensorflow as tf

input1 = tf.constant(3.)
input2 = tf.constant(2.) 
input3 = tf.constant(1.) 

add = tf.add(input1, input2)
mul = tf.multiply(input1, add)

with tf.Session() as sess:
    result = sess.run([mul, add])
    print(result)
```

```python
# feed 传入值
i1 = tf.placeholder(tf.float32)
i2 = tf.placeholder(tf.float32)
output = tf.multiply(i1, i2)

with tf.Session() as sess:
    print(sess.run(output, feed_dict={i1:7,i2:8}))
```

```python
# 简单案例 
import tensorflow as tf
import numpy as np
```

```python
x_data = np.random.rand(100)
y_data = x_data * 0.1 + 0.2

##
b = tf.Variable(10.0)
w = tf.Variable(-10.0)
y = w * x_data + b

loss = tf.reduce_mean(tf.square(y - y_data))

optimizer = tf.train.GradientDescentOptimizer(0.2)

train = optimizer.minimize(loss)

init = tf.global_variables_initializer()

with tf.Session() as sess:
    sess.run(init)
    for step in range(201):
        sess.run(train)
        if step % 20 == 0:
            print(sess.run([w, b]))
```
